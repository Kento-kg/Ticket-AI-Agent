```python
from langchain.tools import tool

@tool
def squre_root(x:float) -> float:
    """caclulate..."""
    return x ** 0.5

def square(x:float) -> float:
    """calculate..."""
    return x ** 2
```

```python
from langchain.agents import create_agent

subagent1 = create_agent(
    model='claude-sonnet-4-6',
    tools=[square_root]
)

subagent2 = create_agent(
    model='claude-sonnet-4-6',
    tools=[square]
)
```
